```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
plt.style.use('seaborn-white')
#matplotlib.rc('xtick', labelsize=20) 
#matplotlib.rc('ytick', labelsize=20)
matplotlib.rc('font',family='Times New Roman')
matplotlib.rcParams['font.size'] = 20
```

```python
mnist=np.loadtxt('mnist/mnist_latent.txt', skiprows=1)
cifar10=np.loadtxt('cifar10/cifar10_latent.txt', skiprows=1)
print (mnist.shape)
```

```python
f, axarr = plt.subplots(2, sharex=True, figsize=(12, 6))
plt.subplot(2,1,1)
plt.plot(mnist[:,0], mnist[:,1], 'C3o-', label='Val Loss')
plt.grid()
plt.ylabel('Error')
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5))
plt.title("MNIST error as a function of latent space dimensionality")

plt.subplot(2,1,2)
plt.plot(mnist[:,0], mnist[:,2], 'C0o-', label='Test MSE')
plt.xlabel('Latent dims')
plt.ylabel('Error')
plt.tight_layout()
plt.grid()
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5))
plt.tick_params(which='both', width=2)
plt.tick_params(which='major', length=7)
plt.tick_params(which='minor', length=5)
plt.savefig('mnist_error.png', dpi=300)
```
