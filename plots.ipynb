```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
plt.style.use('seaborn-white')
#matplotlib.rc('xtick', labelsize=20) 
#matplotlib.rc('ytick', labelsize=20)
matplotlib.rc('font',family='Times New Roman')
matplotlib.rcParams['font.size'] = 20
```

```python
mnist1=np.loadtxt('mnist/mnist_latent_1e_4.txt', skiprows=1)
mnist2=np.loadtxt('mnist/mnist_latent_2e_4.txt', skiprows=1)
mnist3=np.loadtxt('mnist/mnist_latent_5e_4.txt', skiprows=1)
mnist4=np.loadtxt('mnist/mnist_latent_50e_4.txt', skiprows=1)
cifar10=np.loadtxt('cifar10/cifar10_latent.txt', skiprows=1)
print (mnist1.shape)
```

```python
f, axarr = plt.subplots(2, sharex=True, figsize=(12, 6))
plt.subplot(2,1,1)
plt.plot(mnist1[:,0], mnist1[:,1], 'C1o-', label='1e-4')
plt.plot(mnist2[:,0], mnist2[:,1], 'C2o-', label='2e-4')
plt.plot(mnist3[:,0], mnist3[:,1], 'C3o-', label='5e-4')
plt.plot(mnist4[:,0], mnist4[:,1], 'C4o-', label='50e-4')
plt.grid()
plt.ylabel('Val Loss')
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5))
plt.title("MNIST error as a function of latent space dimensionality")
plt.xlim(xmax=256, xmin=8)
plt.ylim(ymax=0.15)

plt.subplot(2,1,2)
plt.plot(mnist1[:,0], mnist1[:,2], 'C1o-', label='1e-4')
plt.plot(mnist2[:,0], mnist2[:,2], 'C2o-', label='2e-4')
plt.plot(mnist3[:,0], mnist3[:,2], 'C3o-', label='5e-4')
plt.plot(mnist4[:,0], mnist4[:,2], 'C4o-', label='50e-4')
plt.xlabel('Latent dims')
plt.ylabel('MSE Error')
plt.tight_layout()
plt.grid()
plt.xlim(xmax=256, xmin=8)
plt.ylim(ymax=0.025)
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5))
plt.tick_params(which='both', width=2)
plt.tick_params(which='major', length=7)
plt.tick_params(which='minor', length=5)
plt.savefig('mnist_error.png', dpi=300)
```
